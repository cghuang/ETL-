```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.skbank.com.tw/2006_credit/credit_08.asp'
res = requests.get(url)
res.encoding = 'big5'
soup = BeautifulSoup(res.text)

div = soup.select('b')

div2 = soup.select('td')

k=1
for skb in div:
    #print 'Q{} : '.format(k) + skb.text.strip().split(' ')[0].encode('unicode-escape')
    print 'Q{} : '.format(k) + skb.text.strip().split(' ')[0]
    print 'len = '+str(len(skb.text.strip().split(' ')[0]))
    k +=1

    
k=1    
for skb2 in div2:
    print 'A{} : '.format(k)
    print skb2.text.strip()
    k +=1
```
