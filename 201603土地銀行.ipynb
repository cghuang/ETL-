# 存款業務類

```python
#存款業務

import requests
from bs4 import BeautifulSoup
url = 'http://www.landbank.com.tw/webpad/webpad.aspx?EpfJdId9UuD72hs5%2fo3zjU%2bbAZuwjvVb'
res = requests.get(url)
soup = BeautifulSoup(res.text)


div = soup.select('.title02')

div2 = soup.select('.title_com')

k=1
for land in div:
    print 'Q{} : '.format(k) + land.text.strip().split(' ')[0]
    k +=1

    
k=1    
for land2 in div2:
    print 'A{} : '.format(k)
    print land2.text.strip()
    k +=1
```
